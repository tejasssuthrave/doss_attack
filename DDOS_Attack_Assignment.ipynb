## Load the data



```python
import pandas as pd

df = pd.read_csv('/content/Friday-WorkingHours-Afternoon-DDos.pcap_ISCX.csv')
display(df.head())
display(df.shape)
```

## Explore the data


```python
display(df.columns)
```

```python
df.columns = df.columns.str.strip()
df.info()
display(df.describe())
display(df.isnull().sum())
display(df['Label'].value_counts())
```

## Preprocess the data



```python
import numpy as np

# Identify columns with constant values
constant_columns = [col for col in df.columns if df[col].nunique() == 1]
df_cleaned = df.drop(columns=constant_columns)

# Identify and handle infinite values
df_cleaned.replace([np.inf, -np.inf], np.nan, inplace=True)
df_cleaned.dropna(inplace=True)

# Separate features (X) and target (y)
X = df_cleaned.drop(columns=['Label'])
y = df_cleaned['Label']

display(X.head())
display(y.head())
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Encode the categorical target variable
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Scale the numerical features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

display(X_scaled[:5])
display(y_encoded[:5])
```

## Prepare data for modeling



```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_encoded, test_size=0.2, random_state=42)

display(X_train.shape)
display(X_test.shape)
display(y_train.shape)
display(y_test.shape)
```

## Choose and train a machine learning model


```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
```

## Evaluate the model



```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-score: {f1:.4f}")
```

## Demonstrate ddos detection



```python
# Select a few instances (assuming the first few are BENIGN and some later ones are DDoS based on the value counts)
benign_samples = df_cleaned[df_cleaned['Label'] == 'BENIGN'].head()
ddos_samples = df_cleaned[df_cleaned['Label'] == 'DDoS'].head()
selected_instances = pd.concat([benign_samples, ddos_samples])

# Separate features and labels for selected instances
selected_features = selected_instances.drop(columns=['Label'])
selected_labels = selected_instances['Label']

# Preprocess the selected instances: scale the numerical features using the fitted scaler
selected_features_scaled = scaler.transform(selected_features)

# Predict the labels using the trained model
predicted_labels_encoded = model.predict(selected_features_scaled)
predicted_labels = label_encoder.inverse_transform(predicted_labels_encoded)

# Display the original selected instances and their predicted labels
results_df = selected_instances.copy()
results_df['Predicted_Label'] = predicted_labels

display(results_df[['Label', 'Predicted_Label']])
```

```python
# ===========================================
# ✅ CONTINUATION: Model Validation, Leakage Check, and Saving (Fixed)
# ===========================================

# Import required libraries
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import joblib

from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.impute import SimpleImputer
from sklearn.metrics import (
    classification_report, confusion_matrix, accuracy_score,
    roc_auc_score, roc_curve
)
from sklearn.inspection import permutation_importance

# -------------------------------------------
# 📘 Step 1: Load Dataset
# -------------------------------------------
DATA_PATH = "/content/Friday-WorkingHours-Afternoon-DDos_ISCX.csv"  # update path if needed
RANDOM_STATE = 42

print("Loading dataset...")
df = pd.read_csv(DATA_PATH)
print("Dataset loaded successfully with shape:", df.shape)

# -------------------------------------------
# 📘 Step 2: Identify Target Column Automatically
# -------------------------------------------
target_candidates = ['Label', 'label', 'Type', 'type', 'Attack', 'attack', 'Class', 'class']
target = None
for col in target_candidates:
    if col in df.columns and df[col].nunique() > 1:
        target = col
        break
if target is None:
    target = df.columns[-1]

print(f"Target column identified as: {target}")

# -------------------------------------------
# 📘 Step 3: Data Cleaning & Preprocessing
# -------------------------------------------
drop_cols = [c for c in ['Flow ID', 'flow_id', 'ID', 'id', 'Src IP', 'Dst IP'] if c in df.columns]
X = df.drop(columns=drop_cols + [target])
y = df[target]

# Encode labels if categorical
if y.dtype == 'object' or str(y.dtype).startswith('category'):
    le = LabelEncoder()
    y = le.fit_transform(y.astype(str))
    print("Encoded target classes:", list(le.classes_))
else:
    le = None

# Convert all columns to numeric (non-numeric become NaN)
X = X.apply(pd.to_numeric, errors='coerce')

# 🔧 Fix: Replace infinities with NaN before imputation
X.replace([np.inf, -np.inf], np.nan, inplace=True)

# Impute missing values with median
imputer = SimpleImputer(strategy='median')
X = pd.DataFrame(imputer.fit_transform(X), columns=X.columns)

# Standardize numerical features for model stability
scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

# -------------------------------------------
# 📘 Step 4: Train-Test Split (80-20)
# -------------------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=RANDOM_STATE, stratify=y
)
print("Train-Test Split Completed:")
print("Training Samples:", X_train.shape[0], "| Testing Samples:", X_test.shape[0])

# -------------------------------------------
# 📘 Step 5: Check for Data Leakage
# -------------------------------------------
duplicates = df.duplicated().sum()
print(f"Total duplicate rows in dataset: {duplicates}")

overlap = len(set(map(tuple, X_train.values)) & set(map(tuple, X_test.values)))
print(f"Overlapping rows between Train/Test: {overlap}")

corr_with_target = {}
for col in X_train.columns:
    try:
        corr = np.corrcoef(X_train[col], y_train)[0, 1]
    except Exception:
        corr = 0
    corr_with_target[col] = 0 if np.isnan(corr) else abs(corr)

leak_features = [f for f, v in corr_with_target.items() if v > 0.9999]
if leak_features:
    print("⚠️ Potential Leakage Features Found:", leak_features)
else:
    print("✅ No highly correlated leakage features detected.")

# -------------------------------------------
# 📘 Step 6: K-Fold Cross Validation
# -------------------------------------------
rf = RandomForestClassifier(n_estimators=200, random_state=RANDOM_STATE, n_jobs=-1)
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)
cv_scores = cross_val_score(rf, X_scaled, y, cv=cv, scoring='accuracy', n_jobs=-1)

print("\nK-Fold Cross Validation Results:")
print("Accuracy scores:", cv_scores)
print("Mean Accuracy: {:.4f} | Std: {:.4f}".format(cv_scores.mean(), cv_scores.std()))

# -------------------------------------------
# 📘 Step 7: Train Final Model on Training Data
# -------------------------------------------
rf_final = RandomForestClassifier(n_estimators=200, random_state=RANDOM_STATE, n_jobs=-1)
rf_final.fit(X_train, y_train)
y_pred = rf_final.predict(X_test)

# -------------------------------------------
# 📘 Step 8: Model Evaluation
# -------------------------------------------
acc = accuracy_score(y_test, y_pred)
print("\nModel Evaluation on Test Set:")
print("Accuracy:", round(acc, 4))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

if len(np.unique(y_test)) == 2:
    y_prob = rf_final.predict_proba(X_test)[:, 1]
    auc = roc_auc_score(y_test, y_prob)
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    plt.figure(figsize=(6, 4))
    plt.plot(fpr, tpr, label=f"AUC = {auc:.4f}")
    plt.plot([0, 1], [0, 1], 'k--', lw=0.8)
    plt.title("ROC Curve - Random Forest")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.legend()
    plt.grid(True)
    plt.show()

# -------------------------------------------
# 📘 Step 9: Permutation Importance
# -------------------------------------------
print("\nCalculating Permutation Importances...")
perm_importance = permutation_importance(
    rf_final, X_test, y_test, n_repeats=10, random_state=RANDOM_STATE, n_jobs=-1
)
perm_df = pd.DataFrame({
    'Feature': X_test.columns,
    'Importance_Mean': perm_importance.importances_mean,
    'Importance_Std': perm_importance.importances_std
}).sort_values(by='Importance_Mean', ascending=False)

print("\nTop 15 Important Features (Permutation Importance):")
display(perm_df.head(15))

# -------------------------------------------
# 📘 Step 10: Save Model and Artifacts
# -------------------------------------------
MODEL_PATH = "/content/ddos_rf_model.pkl"
ARTIFACTS_PATH = "/content/ddos_preprocessing.pkl"

joblib.dump(rf_final, MODEL_PATH)
joblib.dump({
    'imputer': imputer,
    'scaler': scaler,
    'label_encoder': le,
    'feature_columns': X.columns.tolist()
}, ARTIFACTS_PATH)

print("\n✅ Model and preprocessing artifacts saved successfully.")
print("Model path:", MODEL_PATH)
print("Artifacts path:", ARTIFACTS_PATH)
print("\n🎯 All steps completed without errors.")
```
